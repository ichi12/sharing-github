```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('gambar1.jpg')
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize =
(20, 20))

for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'gray')
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2,
figsize = (20, 20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
img_copy = img.copy()

cv2.rectangle(img_copy, pt1 = (800, 470), pt2 = (980,
530),
color = (255, 0, 0), thickness = 5)
plt.imshow(img_copy)
```

```python
cv2.circle(img_copy, center = (950, 50), radius = 50,
color = (0, 0, 255), thickness = 5)
plt.imshow(img_copy)
```

```python
cv2.putText(img_copy, text = "tenki no ko",
org = (250, 250),
fontFace = cv2.FONT_HERSHEY_DUPLEX,
fontScale = 2,
color = (0, 255, 0),
thickness = 2,
lineType = cv2.LINE_AA)
plt.imshow(img_copy)
```

```python
import cv2
import numpy as np

drawing = False
ix = -1
iy = -1

def draw_rectangle(event, x, y, flags, param):
    
    global ix, iy, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img, pt1=(ix, iy), pt2=(x, y),
                          color = (87, 184, 237), thickness = -1)

        elif event == cv2.EVENT_LBUTTONUP:
            drawing = False
            cv2.rectangle(img, pt1=(ix, iy), pt2=(x, y),
                          color = (87, 184, 237), thickness =-1)

img = cv2.imread('gambar.jpg')

cv2.namedWindow(winname = 'my_drawing')
cv2.setMouseCallback('my_drawing', draw_rectangle)

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(10) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
